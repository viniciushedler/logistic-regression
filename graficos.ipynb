```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
base1_tamanho = 40
peso0 = np.random.normal(loc=80, scale=7, size=base1_tamanho)
peso1 = np.random.normal(loc=100, scale=10, size=base1_tamanho)
peso = np.concatenate((peso0, peso1))
peso
```

```python
obesidade0 = np.full((base1_tamanho,),0)
obesidade1 = np.full((base1_tamanho,),1)
obesidade = np.concatenate((obesidade0, obesidade1))
obesidade
```

```python
df_obesidade = pd.DataFrame()
df_obesidade["peso"] = peso
df_obesidade["obesidade"] = obesidade
df_obesidade.head()
```

```python
grafico1 = sns.lmplot(x="peso", y="obesidade", data=df_obesidade, y_jitter=.02, fit_reg=False)
```

```python
grafico2 = sns.lmplot(x="peso", y="obesidade", data=df_obesidade, logistic=False, y_jitter=.02)
```

```python
grafico3 = sns.lmplot(x="peso", y="obesidade", data=df_obesidade, logistic=True, y_jitter=.02)
```

```python
df_obesidade["previsto"] = np.where(df_obesidade.peso >= 90, 1, 0)
df_obesidade
```

```python
grafico4 = sns.lmplot(x="peso", y="obesidade", data=df_obesidade, logistic=True, y_jitter=.02, color="obesidade")
```
